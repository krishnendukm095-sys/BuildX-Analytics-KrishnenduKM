```python
import pandas as pd
```

```python
df=pd.read_csv("/zomatoes.csv")
```

```python
print("First 5 rows:")
print(df.head())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df.dropna(subset=["restaurant name"],inplace=True)
```

```python
df["restaurant type"]=df["restaurant type"].fillna("Unknown")
```

```python
df["cuisines type"] = df["cuisines type"].fillna("Unknown")
```

```python
if "rate (out of 5)" in df.columns:
    df["rate (out of 5)"] = df["rate (out of 5)"].fillna(
        df["rate (out of 5)"].median()
    )
```

```python
if "avg cost (two people)" in df.columns:
    df["avg cost (two people)"] = df["avg cost (two people)"].fillna(
        df["avg cost (two people)"].median()
    )
```

```python
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.title()
    )

if "online_order" in df.columns:
    df["online_order"] = df["online_order"].str.lower()

if "table booking" in df.columns:
    df["table booking"] = df["table booking"].str.lower()
```

```python
if "rate (out of 5)" in df.columns:
    df["rate (out of 5)"] = pd.to_numeric(
        df["rate (out of 5)"],
        errors="coerce"
    )

if "num of ratings" in df.columns:
    df["num of ratings"] = pd.to_numeric(
        df["num of ratings"],
        errors="coerce"
    )

if "avg cost (two people)" in df.columns:
    df["avg cost (two people)"] = pd.to_numeric(
        df["avg cost (two people)"],
        errors="coerce"
    )
```

```python
df.rename(columns={
    "restaurant name": "restaurant_name",
    "restaurant type": "restaurant_type",
    "rate (out of 5)": "rating",
    "num of ratings": "num_ratings",
    "avg cost (two people)": "avg_cost_two_people",
    "table booking": "table_booking",
    "cuisines type": "cuisines_type"
}, inplace=True)
```

```python
if "Unnamed: 0" in df.columns:
    df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
def rating_category(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Very Good"
    elif rating >= 3.0:
        return "Good"
    else:
        return "Average"

df["rating_category"] = df["rating"].apply(rating_category)
```
